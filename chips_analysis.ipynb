```python


# Convert Excel serial date (43390) to datetime
df['DATE'] = pd.to_datetime(df['DATE'], origin='1899-12-30', unit='D')

# Format as d/m/y (e.g., 25/10/2018)
df['DATE'] = df['DATE'].dt.strftime('%d/%m/%Y')

print(df['DATE'])
```

```python
df['GRAM'] = df['PROD_NAME'].str.extract(r'(\d+)[gG]')[0].astype(int)
```
